#### Distribution of laptop(Top 10)

```python
df["brand"].value_counts().reset_index().head(10)
```

```python
df["brand"].value_counts().reset_index().head(10).plot(x="brand",kind="bar")
```

##### Distribution of Laptop Price

```python
df["Price"].value_counts().sort_values()
```

```python
sns.histplot(df['Price'],bins=10)
plt.title('Distribution of Laptop Prices')
plt.show()
```

##### distribution of Rating with Price

```python
plt.scatter(x=df["Rating"],y=df["Price"])
plt.title("Distribution of Rating with Price")
plt.show()
```

##### Top 10 Expensive laptop

```python
df.sort_values(by="Price",ascending=False).head(10)
```

##### secondary storage type

```python
df["secondary_storage_type"].value_counts()
```

```python
sns.histplot(df["secondary_storage_type"])
plt.title("Distribution of secondary storage type")
plt.show()
```

##### Laptops which has SSD card as a secondary storage

```python
df.loc[(df["secondary_storage_type"]=="SSD"),['brand', 'Model', 'Price', 'Rating','is_touch_screen']].sort_values(by="Price",ascending=False)
```

##### No. of years of warranty given

```python
warrenty_counts=df["year_of_warranty"].value_counts()
warrenty_counts
```

```python
df["year_of_warranty"].value_counts().plot(kind="pie",autopct="%1.1f%%")
```

##### Types of operating system in laptop

```python
df["OS"].value_counts()
```

```python
df["OS"].value_counts().plot(kind="bar",title="Type of Operating systems")
```

```python
df["primary_storage_capacity"].value_counts().plot(kind="bar",title="Types of primary storage")
```

##### Types of processor Brands

```python
df["processor_brand"].value_counts()
```

```python
sns.countplot(data=df, x='processor_brand')
plt.title('Distribution of Processor Brands')
plt.show()
```

```python
df.groupby("processor_tier")["brand"].value_counts().reset_index()
```

```python
df["processor_tier"].value_counts()
```

```python
sns.countplot(data=df, x='processor_tier')
plt.title('Distribution of Processor Tiers')
plt.xlabel('Processor Tier')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
total_price_brand=df.groupby("brand")["Price"].sum().reset_index().sort_values(by="Price",ascending=False)
total_price_brand
```

```python
plt.figure(figsize=(10,8))
plt.bar(total_price_brand['brand'], total_price_brand['Price'])
plt.title('Total Price of Laptops by Brand')
plt.xticks(rotation=90)
plt.show()
```

```python
df.groupby("resolution_width")["resolution_height"].sum().reset_index()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['display_size'])
plt.title('Distribution of Display Sizes')
plt.show()
```
